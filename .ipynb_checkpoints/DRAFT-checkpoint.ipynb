Условие:
Вам дана перестановка чисел от 1 до n: p1, p2, ..., pn (1 <= pi <= n), все числа различны).

Для каждого индекса i найдите сумму si стоящих левее этой позиции чисел и меньших pi.

Примеры:

1. 
n = 5
pi = [1, 3, 5, 4, 2]
Вывод: [0, 1, 4, 4, 1]


2. 
n = 6
pi = [6, 5, 4, 3, 2, 1]
Вывод: [0, 0, 0, 0, 0, 0]


3. 
n = 10
pi = [9, 5, 7, 10, 1, 4, 2, 3, 8, 6]
Вывод: [0, 0, 5, 21, 0, 1, 1, 3, 22, 15]

```python
def check_int(l, n):
    for i in l:
        if not i.isnumeric():
            print('Вводите числа')
            break
        if len(l) != n:
            print(f'Количество значений должно быть {n}')
            break
        
    return list(map(int,l))
try:
    n = int(input())
except ValueError:
    print('Должно быть целым числом')
pi = check_int(list(map(str, input().split(' '))), n)
```

```python
import numpy as np
n = 10000
pi = np.random.randint(0,1000,10000)
```

```python
%%time
res = []
for i in range(n):
    s = 0    
    for j in range(i+1):
        if pi[j] < pi[i]:
            s += pi[j]
    res.append(s)
            
```

```python
n = 10
pi = [9, 5, 7, 10, 1, 4, 2, 3, 8, 6]
# Вывод: [0, 0, 5, 21, 0, 1, 1, 3, 22, 15]
```

```python
def func(pi, n):
    res = []
    for i in range(n):
        res.append(sum(list(filter(lambda x:x<pi[i], pi[:i]))))
    return res
```

```python
%%time
func(pi,n)
```
